```python
GrfFile = 'E:/first_ise/AAedion/outputfile.txt'
DmFile = 'E:/first_ise/AAedion/outputfile1.txt'
GrfConsole = open(GrfFile)
DmConsole = open(DmFile)

resultList = []

lineGrfConsole = GrfConsole.readline()
while lineGrfConsole:
    resultList.append(lineGrfConsole)
    lineGrfConsole = GrfConsole.readline()

lineDmConsole = DmConsole.readline()
while lineDmConsole:
    resultList.append(lineDmConsole)
    lineDmConsole = DmConsole.readline()

resultList.sort()
resultList
```
